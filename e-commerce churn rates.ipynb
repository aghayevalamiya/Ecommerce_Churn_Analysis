```python
def euclidean(row0: dict, row1: dict) -> float:
    distance_def = 0
    for i in row1:
        if not pd.isna(row0[i]):
            distance_def += np.power(row0[i] - row1[i], 2.0)   # (two reasons, 1: make it "positive", 2: when the difference gets bigger, the distance must get "bigger" exponantionaly -- EXTREMELY
    distance_def = np.sqrt(distance_def)
    return distance_def
```

```python
def row(row):
    distance={}
    for i in list(df_without_nulls.index):
        row1 = dict(df_without_nulls.loc[i])
        distance[i] = euclidean(row,row1)
    min_distance = min(distance.items(), key=lambda x: x[1]) 
    min_distance = list(min_distance)
    return min_distance
```

```python
distance = {}
for i in list(df_tenure.index):
    row0 = dict(df_tenure.loc[i])
    distance[i] = row(row0)
distance
```

```python
for i in list(distance.keys()):
    value = int(df.loc[distance[i][0]]['Tenure'])
    df['Tenure'][i] = (value)
df['Tenure'].isnull().sum()
```

```python
outliers = {}
for i in df:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3-q1
    upperbound = q3 + 1.5*iqr
    lowerbound = q1 - 1.5*iqr
    outlier = df[(df[i] > upperbound) | (df[i] < lowerbound)]
    outliers[i] = len(outlier)
    if i == 'Cashback':
        df[i] = df[i].clip(lowerbound,upperbound)
        outliers[i+'_new'] = sum((df[i] > upperbound) | (df[i] < lowerbound))
outliers
```
